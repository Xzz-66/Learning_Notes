```python
# 
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns

# 
iris_sns = sns.load_dataset("iris")

# 
fig, ax = plt.subplots(figsize = (8,6))

sns.histplot(data = iris_sns, x = "sepal_length",
             binwidth = 0.2, ax =ax)
# bins  binwidth 

# 
ax.axvline(x = iris_sns.sepal_length.mean(),
           color = 'r', ls = '--')
# 

# 
mean = iris_sns.sepal_length.mean()
std_dev = iris_sns.sepal_length.std()

# 
ax.axvline(x = mean + std_dev, color='g', ls='--', label='Mean  SD')
ax.axvline(x = mean - std_dev, color='g', ls='--')

# 
ax.legend()

plt.show()
```

```python
# 
fig, ax = plt.subplots(figsize = (8,6))

sns.histplot(data = iris_sns, x="sepal_length",
             hue = 'species', binwidth = 0.2, ax = ax,
             element = "step", stat = 'density')
'''
 stat 'count''probability' 'density' 
 hue = 'species'
'''
```
