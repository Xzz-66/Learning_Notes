# Seaborn 可视化


# 一元特征数据
# 直方图

```python
# 导入包
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns

# 导入鸢尾花数据
iris_sns = sns.load_dataset("iris")

# 绘制鸢尾花长度样本数据直方图
fig, ax = plt.subplots(figsize = (8,6))

sns.histplot(data = iris_sns, x = "sepal_length",
             binwidth = 0.2, ax =ax)
# bins 指定直方图区间数量； binwidth 定义区间宽度

# 纵轴三个选择：频率，概率，概率密度
ax.axvline(x = iris_sns.sepal_length.mean(),
           color = 'r', ls = '--')
# 增加均值位置竖直参考线

# 计算均值和标准差
mean = iris_sns.sepal_length.mean()
std_dev = iris_sns.sepal_length.std()

# 绘制均值加减一个标准差的竖线
ax.axvline(x = mean + std_dev, color='g', ls='--', label='Mean ± SD')
ax.axvline(x = mean - std_dev, color='g', ls='--')

# 添加图例
ax.legend()

plt.show()
```

```python
# 绘制花萼长度样本数据直方图，考虑鸢尾花分类
fig, ax = plt.subplots(figsize = (8,6))

sns.histplot(data = iris_sns, x="sepal_length",
             hue = 'species', binwidth = 0.2, ax = ax,
             element = "step", stat = 'density')
'''
参数 stat 指定纵轴类型，比如'count'对应频数，'probability'对应概率， 'density' 对应概率密度。
指定 hue = 'species'，我们便得到每个类别鸢尾花单独的直方图
'''
```
